```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Reading the data

```python
mat = pd.read_csv('student-mat.csv', sep=';')
mat.head()
```

```python
port = pd.read_csv('student-por.csv', sep=';')
port.head()
```

```python
mat2 = mat
port2 = port
```

## Dropping unwanted columns

```python
mat2 = mat2.drop(columns = ['school','age', 'Medu', 'Fedu','Mjob','Fjob', 'Dalc', 'reason'], axis = 1)
port2 = port2.drop(columns = ['school','age', 'Medu', 'Fedu','Mjob','Fjob', 'Dalc', 'reason'], axis = 1)
```

## Encoding with lambda

```python
mat2['sex'] = mat2.sex.apply(lambda v: 0 if v =='F' else 1 )
port2['sex'] = port2.sex.apply(lambda v: 0 if v =='F' else 1 )
```

```python
mat2['famsize'] = mat2.famsize.apply(lambda v: 0 if v =='LE3' else 1 )
port2['famsize'] = port2.famsize.apply(lambda v: 0 if v =='LE3' else 1 )
```

```python
mat2['address'] = mat2.address.apply(lambda v: 0 if v =='U' else 1)
port2['address'] = port2.address.apply(lambda v: 0 if v =='U' else 1)
```

```python
mat2['Pstatus'] = mat2.Pstatus.apply(lambda v: 0 if v =='T' else 1 )
port2['Pstatus'] = port2.Pstatus.apply(lambda v: 0 if v =='T' else 1 )
```

```python
mat2['absences'] = mat2.absences.apply(lambda v: v/93)
port2['absences'] = port2.absences.apply(lambda v: v/93)
```

```python
mat2['schoolsup'] = mat2.schoolsup.apply(lambda v: 0 if v =='no' else 1 )
port2['schoolsup'] = port2.schoolsup.apply(lambda v: 0 if v =='no' else 1 )
```

```python
mat2['health'] = mat2.health.apply(lambda v: (v-1)/4)
port2['health'] = port2.health.apply(lambda v: (v-1)/4)
```

```python
mat2['famsup'] = mat2.famsup.apply(lambda v: 0 if v =='no' else 1 )
port2['famsup'] = port2.famsup.apply(lambda v: 0 if v =='no' else 1 )
```

```python
mat2['paid'] = mat2.paid.apply(lambda v: 0 if v =='no' else 1 )
port2['paid'] = port2.paid.apply(lambda v: 0 if v =='no' else 1 )
```

```python
mat2['activities'] = mat2.activities.apply(lambda v: 0 if v =='no' else 1 )
port2['activities'] = port2.activities.apply(lambda v: 0 if v =='no' else 1 )
```

```python
mat2['nursery'] = mat2.nursery.apply(lambda v: 0 if v =='no' else 1 )
port2['nursery'] = port2.nursery.apply(lambda v: 0 if v =='no' else 1 )
```

```python
mat2['higher'] = mat2.higher.apply(lambda v: 0 if v =='no' else 1 )
port2['higher'] = port2.higher.apply(lambda v: 0 if v =='no' else 1 )
```

```python
mat2['internet'] = mat2.internet.apply(lambda v: 0 if v =='no' else 1 )
port2['internet'] = port2.internet.apply(lambda v: 0 if v =='no' else 1 )
```

```python
mat2['romantic'] = mat2.romantic.apply(lambda v: 0 if v =='no' else 1 )
port2['romantic'] = port2.romantic.apply(lambda v: 0 if v =='no' else 1 )
```

```python
mat2['Walc'] = mat2.Walc.apply(lambda v: (v-1)/4)
port2['Walc'] = port2.Walc.apply(lambda v: (v-1)/4)
```

```python
mat2['freetime'] = mat2.freetime.apply(lambda v: (v-1)/4)
port2['freetime'] = port2.freetime.apply(lambda v: (v-1)/4)
```

```python
mat2['goout'] = mat2.goout.apply(lambda v: (v-1)/4)
port2['goout'] = port2.goout.apply(lambda v: (v-1)/4)
```

```python
mat2['studytime'] = mat2.studytime.apply(lambda v: (v-1)/3)
port2['studytime'] = port2.studytime.apply(lambda v: (v-1)/3)
```

```python
mat2['traveltime'] = mat2.traveltime.apply(lambda v: (v-1)/3)
port2['traveltime'] = port2.traveltime.apply(lambda v: (v-1)/3)
```

```python
mat2['famrel'] = mat2.famrel.apply(lambda v: (v-1)/4)
port2['famrel'] = port2.famrel.apply(lambda v: (v-1)/4)
```

## Setting the grade threshold

```python
#mat2['Improve'] = mat2.G2-mat2.G1
#port2['Improve'] = port2.G2-port2.G1
mat2.head()
```

```python
mat2['G1'] = mat2.G1.apply(lambda v: 0 if v<12 else 1)
port2['G1'] = port2.G1.apply(lambda v: 0 if v<12 else 1)
mat2['G2'] = mat2.G2.apply(lambda v: 0 if v<12 else 1)
port2['G2'] = port2.G2.apply(lambda v: 0 if v<12 else 1)
mat2['G3'] = mat2.G3.apply(lambda v: 0 if v<12 else 1)
port2['G3'] = port2.G3.apply(lambda v: 0 if v<12 else 1)
#mat2['Improve'] = mat2.Improve.apply(lambda v: 0 if v<0 else 1)
#port2['Improve'] = port2.Improve.apply(lambda v: 0 if v<0 else 1)
```

## One hot encoding

```python
tmp = pd.get_dummies(mat2.guardian,prefix = 'guardian', drop_first=True)
mat2 = pd.concat([mat2, tmp], axis=1 )
mat2 = mat2.drop(columns = ['guardian'], axis = 1)
tmp = pd.get_dummies(port2.guardian,prefix = 'guardian', drop_first=True)
port2 = pd.concat([port2, tmp], axis=1 )
port2 = port2.drop(columns = ['guardian'], axis = 1)
```

```python
tmp = pd.get_dummies(mat2.failures,prefix = 'failures', drop_first=True)
mat2 = pd.concat([mat2, tmp], axis=1 )
mat2 = mat2.drop(columns = ['failures'], axis = 1)
tmp = pd.get_dummies(port2.failures,prefix = 'failures', drop_first=True)
port2 = pd.concat([port2, tmp], axis=1 )
port2 = port2.drop(columns = ['failures'], axis = 1)
```
